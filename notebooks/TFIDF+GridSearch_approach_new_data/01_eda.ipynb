# Numeric columns analysis

```python
print(f"Procent prawdziwych newsów {sum(df['target']==1)*100/df.shape[0]:.2f}%")
print(f"Procent fałszywych newsów {sum(df['target']==0)*100/df.shape[0]:.2f}%")

plt.figure(figsize=(8,4))
sns.countplot(x='target', data=df,palette='coolwarm')
plt.title('Rozkład klas')
plt.xticks([0, 1], ['Fake', 'Real'])
plt.xlabel('Etykieta')
plt.ylabel('Liczba wpisów')
plt.show()
```

```python
def analyze_numeric_columns(df, num_columns,target_column):
    display(df[num_columns].describe())

    corrs = df[num_columns+ [target_column]].corr()[target_column].sort_values(ascending=False)
    print("Korelacje z targetem:")
    display(corrs.sort_values(ascending=False))

    for col in num_columns:
        plt.figure(figsize=(7, 4))
        sns.boxplot(hue=target_column, y=col, data=df, palette="Set2")
        plt.title(f"Rozkład wartości cechy '{col}' względem  '{target_column}'", fontsize=14)
        plt.xlabel("Klasa (target)")
        plt.ylabel(f"Wartość cechy: {col}")
        plt.tight_layout()
        plt.show()


analyze_numeric_columns(df, num_columns, 'target')
```
